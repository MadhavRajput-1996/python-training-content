## Python operators provide a rich set of tools for performing operations on variables and values. 

```python
# Arithmetic operators
x = 10
y = 3
print(x + y)   # output: 13
print(x - y)   # output: 7
print(x * y)   # output: 30
print(x / y)   # output: 3.3333333333333335
print(x % y)   # output: 1
print(x ** y)  # output: 1000
print(x // y)  # output: 3

# Comparison operators
print(x == y)  # output: False
print(x != y)  # output: True
print(x < y)   # output: False
print(x > y)   # output: True
print(x <= y)  # output: False
print(x >= y)  # output: True

# Logical operators
a = True
b = False
print(a and b) # output: False
print(a or b)  # output: True
print(not a)   # output: False

# Assignment operators
z = 5
z += 3
print(z)       # output: 8

# Membership operators
a = [1, 2, 3, 4, 5]

# ***************** in operator
print(3 in a)  # Output: True
print(6 in a)  # Output: False

# ***************** not in operator
print(3 not in a)  # Output: False
print(6 not in a)  # Output: True

# Identity operators
list1 = [1, 2, 3]
list2 = [1, 2, 3]
print(list1 is list2)    # output: False
print(list1 is not list2)# output: True

# *****************
a = [1, 2, 3]
b = [1, 2, 3]
c = a
print(a is b)  
print(a is c)  

# ******************
print(a is not b) 
print(a is not c)  

# Bitwise Operators

a = 10  # Binary: 1010
b = 4   # Binary: 0100

# Bitwise AND
print(a & b)  # Output: 0 (Binary: 0000)

# Bitwise OR
print(a | b)  # Output: 14 (Binary: 1110)

# Bitwise XOR
print(a ^ b)  # Output: 14 (Binary: 1110)

# Bitwise NOT
print(~a)  # Output: -11 (Binary: -1011)

# Bitwise left shift
print(a << 1)  # Output: 20 (Binary: 10100)

# Bitwise right shift
print(a >> 1)  # Output: 5 (Binary: 0101)




```

#### Ternary Operator
- The ternary operator in Python allows you to evaluate an expression based on a condition in a single line of code. It is a concise way to perform conditional assignments

- syntax

*<expression_if_true> if <condition> else <expression_if_false>*

```python
# Example 1: Assigning a value based on a condition
x = 10
y = 20
max_value = x if x > y else y
print(max_value)  # Output: 20

# Example 2: Using ternary operator in a print statement
a = 5
b = 10
print("a is greater" if a > b else "b is greater")  # Output: b is greater
```

### The del keyword is used to remove references to objects, which may cause the objects to be deleted or garbage-collected.

```python
# Deleting elements from a list
 
my_list = [1, 2, 3, 4, 5]
del my_list[2]  # Deletes the element at index 2
print(my_list)  # Output: [1, 2, 4, 5]

# Deleting a variable or object

x = 10
del x  # Deletes variable x 


# Deleting elements from a dictionary

my_dict = {'a': 1, 'b': 2, 'c': 3}
del my_dict['b']  # Deletes the key 'b' along with its value
print(my_dict)  # Output: {'a': 1, 'c': 3}


# Deleting entire objects

class MyClass:
    x = 10

obj = MyClass()
del obj  # Deletes the entire object obj
```
